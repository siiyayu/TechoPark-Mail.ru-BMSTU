```python
class URLTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        print('CustomTransformer initialization')
    #Return self, nothing else to do here
    def fit( self, X, y = None):
        return self 
    
    def transform(self, X, y = None):
        return np.array(X.apply(lambda x: int('porn' in x or 'порно' in x))).reshape(len(X), 1)

class ColumnExtractor(TransformerMixin):
    
    def __init__(self, column_name):
        self.column_name = column_name
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X[self.column_name]
```

```python
estimators = [
     ('rf1', RandomForestClassifier(n_estimators=100, max_features='sqrt', n_jobs=-1, verbose=True, class_weight='balanced')),
     ('rf2', RandomForestClassifier(n_estimators=500, max_features='sqrt', min_samples_leaf=5, n_jobs=-1, verbose=True, class_weight='balanced')),
     ('rf3', RandomForestClassifier(n_estimators=50, min_samples_leaf=10, n_jobs=-1)),
     ('rf4', RandomForestClassifier(n_estimators=30, n_jobs=-1)),
#      ('rf5', RandomForestClassifier(n_estimators=20,  min_samples_leaf=5, n_jobs=-1)),
#      ('rf6', RandomForestClassifier(n_estimators=20, n_jobs=-1)),
#      ('rf7', RandomForestClassifier(n_estimators=50, n_jobs=-1, class_weight='balanced')),
#      ('rf8', RandomForestClassifier(n_estimators=150, n_jobs=-1)),
#      ('rf9', RandomForestClassifier(n_estimators=90, n_jobs=-1)),
#      ('rf10', RandomForestClassifier(n_estimators=40, n_jobs=-1)),
#      ('rf11', RandomForestClassifier(n_estimators=100, n_jobs=-1)),
#      ('rf12', RandomForestClassifier(n_estimators=100, n_jobs=-1, min_samples_leaf=7)),

     ('svc1', LinearSVC(C=0.1)),
     ('svc2', LinearSVC(C=0.5)),
     ('mnb', MultinomialNB()),
     ('logreg1', LogisticRegression(penalty='l2', C=0.4, n_jobs=-1)),
     ('logreg2', LogisticRegression(penalty='l1', C=0.1, solver='liblinear', n_jobs=-1)),
     ('logreg3', LogisticRegression(penalty='l2', C=0.7, n_jobs=-1)),
     ('logreg4', LogisticRegression(penalty='l1', C=0.8, solver='liblinear', n_jobs=-1))


]
```

```python
pipeline = Pipeline([
    (
        'features', 
        FeatureUnion([
            (
                'title', 
                Pipeline([
                    ('extractor', ColumnExtractor('title')),
                    (
                        'vectorizer', 
                        TfidfVectorizer(
                        ngram_range=(3, 4), lowercase=True)
                    )
                ])
            ),
            (
                'url',
                Pipeline([
                    ('extractor', ColumnExtractor('url')),
                    (
                     'vectorizer', TfidfVectorizer(analyzer='char', ngram_range=(3, 4))   
                    )
                ])
            )
        ])
    ),
#     ('stacking_clf', StackingClassifier(n_jobs=-1, estimators=estimators, final_estimator=XGBClassifier(learning_rate=0.1, n_estimators=50, n_jobs=-1), cv=5))
      ('clf', LogisticRegression(class_weight='balanced', n_jobs=-1))
])
```

```python
parameters = {
              'clf__C': [0.9, 1],
              'features__url__vectorizer__ngram_range': [(3, 4)],
              'features__title__vectorizer__ngram_range': [(1, 1)],
              'features__title__vectorizer__min_df': [1, 3],
              'features__title__vectorizer__max_df': [0.8],
              'features__url__vectorizer__min_df': [1, 3],
              'features__url__vectorizer__max_df': [1]
}
gs_pipeline = GridSearchCV(pipeline, parameters, cv=3)
```

```python
%%time
gs_pipeline.fit(train_df, y_train)
```

```python
gs_pipeline.best_params_
```

```python
test_df["target"] = gs_pipeline.predict(test_df).astype(bool)

test_df[["id", "target"]].to_csv("ml_baseline.csv", index=False)

!cat ml_baseline.csv | head
```
